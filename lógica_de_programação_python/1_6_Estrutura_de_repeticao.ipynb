**Estrutura de repetição**

# For

```python
# for range
for contador in range(10): # Inicia um loop que irá iterar 10 vezes. A função range(10) gera números de 0 até 9. 'contador' recebe cada um desses valores
   print(contador) # Exibe o valor atual de 'contador' em cada iteração (0, 1, 2, ..., 9)
```

```python
# for range + break
for contador in range (1000): # Inicia um loop que teoricamente iria de 0 a 999
  print(contador) # Exibe o valor de 'contador'
  if contador == 2: # Testa a condição: se o valor atual de 'contador' for igual a 2...
    break # ... interrompe Imediatamente a execução de todo o laço 'for', independentemente de ter chegado ao fim do 'range(1000)'
```

```python
lista = ["teste1", "teste2"] # Cria uma lista com duas strings
```

```python
# for simples
for val in lista: # Itera sobre os ELEMENTOS da lista. Na primeira volta, 'val' é "teste1"; na segunda, é "teste2"
  print(val) # Exibe o valor do elemento atual ('teste1', depois 'teste2'
```

```python
# for com enumerate
for i, val in enumerate(lista): # Itera sobre a lista, mas usa a função 'enumerate' para obter o ÍNDICE (i) e o VALOR (val) simultaneamente
  print(i, val) # Exibe o índice (0, 1) e o valor correspondente ('teste1', 'teste2')
```

# For com operadores compostos

```python
lista2 = [2,3,4] # Cria uma nova lista
```

```python
for val in lista2: # Itera sobre os elementos da lista2
  val *= 2 # Multiplica o valor atual de 'val' por 2, usando o operador composto '*=' (ex: 2*2=4, 3*2=6, 4*2=8). **Importante:** Este cálculo altera apenas a variável 'val' DENTRO do loop, não o valor na 'lista2' original
  print(val) # Exibe o resultado da multiplicação (4, 6, 8)
```

# While

```python
val = 1 # Inicializa a variável de controle 'val' com 1
while (val < 6): # Testa a condição: Enquanto 'val' for menor que 6...
  val *=2 # ... multiplica 'val' por 2 em cada repetição (1*2=2, 2*2=4, 4*2=8). Essa linha garante que o loop termine em algum momento
  print(val) # Exibe o novo valor de 'val' a cada iteração (2, 4, 8). Quando 'val' se torna 8, a condição 'val < 6' se torna False e o loop para
```

```python
# while com else
val = 4 # Inicializa a variável 'val' com 4
while val < 10: # Testa a condição: Enquanto 'val' for menor que 10... (4*2=8, 8*2=16)
  val *=2 # Multiplica 'val' por 2
else: # O bloco 'else' do while é executado SOMENTE se o loop terminar naturalmente (ou seja, a condição se tornar False). Ele NÃO é executado se o loop for interrompido por um 'break'
  print('valor de val é:', val) # Quando o loop para (porque val=16), esta linha é executada. Resultado: 'valor de val é: 16'
```
