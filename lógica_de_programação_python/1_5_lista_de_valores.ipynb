## **Listas,Tuplas e Dicionários**

```python
# lista
[1,2,3,4,5] # Uma Lista: Coleção ordenada de itens, MUTÁVEL (pode ser alterada) e definida por colchetes []
```

```python
# tupla
(1,2,3,4,5) # Uma Tupla: Coleção ordenada de itens, IMUTÁVEL (NÃO pode ser alterada) e definida por parênteses ()
```

```python
# dicionário
{"num1":1, "num2":2} # Um Dicionário: Coleção de pares chave-valor, desordenada e definida por chaves {}
```

# Lista

```python
lista = [5, "rodrigo",8.9] # Cria uma lista chamada 'lista' com um inteiro, uma string e um float
```

```python
type(lista) # Verifica o tipo de 'lista', que retorna <class 'list'>
```

```python
lista.append("augusto") # Usa o método '.append()' para ADICIONAR o elemento "augusto" no FINAL da lista
```

```python
print(lista) # Exibe a lista completa: [5, 'rodrigo', 8.9, 'augusto']
print(lista[1]) # Acessa e exibe o elemento no ÍNDICE 1 da lista (Lembre-se: a contagem começa em 0). Resultado: 'rodrigo'
```

# Tupla

```python
tupla = (5, "rodrigo", 8.9) # Cria uma tupla chamada 'tupla'
```

```python
type(tupla) # Verifica o tipo de 'tupla', que retorna <class 'tuple'>
```

```python

tupla = (5, "rodrigo", 8.9) # Recria a tupla original
tupla = tupla + ("augusto",) # Como tuplas são imutáveis, NÃO posso usar .append(). Eu crio uma NOVA tupla combinando a tupla original com uma nova tupla contendo ("augusto",). Note a vírgula depois de "augusto" para indicar que é uma tupla de um único elemento
print(tupla) # Exibe a nova tupla resultante: (5, 'rodrigo', 8.9, 'augusto')
```

```python
print(tupla) # Exibe novamente o conteúdo final da tupla
```

# Dicionário

```python
dicionario = {'nome':'rodrigo','cidade':'São josé dos campos'} # Cria um dicionário. 'nome' e 'cidade' são as chaves (Keys); 'rodrigo' e 'São josé dos campos' são os valores (Values)
```

```python
type(dicionario) # Verifica o tipo, que retorna <class 'dict'>
```

```python
print(dicionario) # Exibe o dicionário completo
```

```python
print(dicionario['nome']) # Acessa e exibe o VALOR associado à CHAVE 'nome'. Resultado: 'rodrigo'
```

```python
dicionario['ultimo_nome'] = 'Rebouças' # ADICIONA ou ATUALIZA um par chave-valor. Como 'ultimo_nome' não existia, ele é adicionado ao dicionário com o valor 'Rebouças'
```

```python
print(dicionario) # Exibe o dicionário atualizado com o novo par chave-valor
```
