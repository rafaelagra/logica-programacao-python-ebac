# Operadores Relacionais

```python
# Menor
10 < 2 # Estou comparando se 10 é menor que 2. Isso é False
```

```python
# menor ou igual
10 <= 2 # Comparando se 10 é menor ou igual a 2. Isso também é False
```

```python
# Maior
10 > 2 # Comparando se 10 é maior que 2. Isso é True
```

```python
# maior ou igual
10>=2 # Comparando se 10 é maior ou igual a 2. Isso é True
```

```python
# igual
7 == 7 # Verificando se 7 é igual a 7. Isso é True
```

```python
# diferente
7 != 7 # Verificando se 7 é diferente de 7. Isso é False
```

# Operadores Lógicos

- And = O operador "E" lógico. Só retorna True se AMBAS as condições forem True

-Or = O operador "OU" lógico. Retorna True se PELO MENOS UMA condição for True

-Not = O operador "NÃO" lógico. Inverte o valor booleano (True vira False e vice-versa)

# Tabela Verdade

And

```python
print (10 > 2 and 4 < 6) # V and V - V. 10>2 é True E 4<6 é True. O resultado é True
print (10 > 2 and 4 > 6) # V and F - F. 10>2 é True E 4>6 é False. Pelo menos um False, o resultado é False
print (10 < 2 and 4 < 6) # F and V - F. 10<2 é False E 4<6 é True. Pelo menos um False, o resultado é False
print (10 < 2 and 4 > 6) # F and F - F. 10<2 é False E 4>6 é False. O resultado é False
```

Or

```python
print (10 > 2 or 4 < 6) # V or V - V. 10>2 é True OU 4<6 é True. O resultado é True
print (10 > 2 or 4 > 6) # V or F - V. 10>2 é True OU 4>6 é False. Como um é True, o resultado é True
print (10 < 2 or 4 < 6) # F or V - V. 10<2 é False OU 4<6 é True. Como um é True, o resultado é True
print (10 < 2 or 4 > 6) # F or F - F. 10<2 é False OU 4>6 é False. O resultado é False
```

# Not

```python
not 10 > 2    # V- F. 10>2 é True, o 'not' inverte para False
```

```python
not 10 < 2      # F - V. 10<2 é False, o 'not' inverte para True
```
