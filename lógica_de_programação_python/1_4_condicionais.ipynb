**Condicional**

```python
# if
if 5 < 10: # Avalia a condição: 5 é menor que 10? Sim, é True
  print("verdade") # Já que a condição acima foi True, este bloco de código, que está identado, será executado
```

```python
# if/else
num1 = 10 # Atribui o valor 10 à variável 'num1'
num2 = 5 # Atribui o valor 5 à variável 'num2'
if num1 > num2: # Avalia a condição principal: 10 é maior que 5? Sim, é True
  print("o num1 é o maior número ") # Já que a condição foi True, este bloco é executado
else: # Se a condição do 'if' tivesse sido False, o programa pularia para o 'else'
  print("o num2 é o maior número") # Este bloco não é executado neste caso, pois o 'if' foi True
```

```python
# if/elif/else
num1 = 50 # Atribui o valor 50 à variável 'num1'
num2 = 10 # Atribui o valor 10 à variável 'num2'
if num1 > num2: # Testa a primeira condição: 50 é maior que 10? Sim, é True
  print("o num1 é o maior número") # Esta linha é executada
elif num1 < num2: # Esta condição (50 < 10) nem é testada/executada, pois o 'if' anterior já foi True
  print("o num2 é o maior número") # Este bloco é pulado
else: # O 'else' só seria executado se TODAS as condições anteriores (if e elif) tivessem sido False
 print("os dois são iguais") # Este bloco é pulado
```

**Condicional com operadores lógicos**

```python
num1 = 10 # Atribui 10 a 'num1'
num2 = 5 # Atribui 5 a 'num2'
num3 = 2 # Atribui 2 a 'num3'
if (num1 > num2) and (num1 > num3): # Testa a condição composta: (10 > 5) AND (10 > 2)?
                                    # Isso é (True) AND (True), que resulta em True
  print("o numero 1 é o maior numero") # A condição composta é True, então esta linha é executada
else: # Se a condição composta fosse False, este bloco seria executado
  print("o num1 não é o maior número") # Este bloco não é executado
```
